### 1. The Histogram:
#### The histogram is used to summarize discrete data, it provides a visual interpretation of numerical data by showing
#### the number of data points that falls within a specified range of values called ("bins/intervals").

#### An example of how the histogram works.

##### Step 1: Creating a dataset of 100 values with a mean of 5 and a
##### SD of 10.
##### Step 2:Creating the histogram with 20 intervals.
##### Step 3: Add labels and titles.
##### Step 4: Display plot.

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(1000)
plt.hist(data, bins=10, color='orange', edgecolor='r')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

### 2.The Scatter plot:
#### The scatter plot is a visual representation of how two variables relate to each other using dots.

```python
from matplotlib import pyplot as plt 
import numpy as np
x=np.array[1,2,3,4]
y=np.array([60,40,90,80])
color=['r','b','purple','brown']
size=[20,60,10,80]
#or plt.scatter(x,y,color='black' to make all the elements have the same color)
plt.scatter(x,y,c=color,s=size)
plt.title('A Scatter plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

#### To compare two data sets using scatter plot.

```python
from matplotlib import pyplot as plt 
import numpy as np
x=np.array([1,2,3,4])
y=np.array([60,40,90,80])
color=['r']

plt.scatter(x,y,c=color,s=size)

x = np.array([5, 6, 7, 8])
y = np.array([12, 10, 12, 5,])
color=['purple']
size=[20,60,10,80]

plt.scatter(x,y,c=color,s=size)
plt.title('A Scatter plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

### 3. The Box plot:
#### A box plot is a graphical representation that displays the distribution of a dataset based on five summary statistics:
#### minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.
#### It is a useful tool for visualizing the spread and central tendency of the data.

#### Example:



```python
import numpy as np
import matplotlib.pyplot as plt
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(0, 2, 100)
data3 = np.random.normal(0, 3, 100)
data=[data1,data2,data3]
plt.boxplot(data, labels=['Data 1', 'Data 2', 'Data 3'], patch_artist=True, notch=True, vert=True)
plt.title('A Box Plot')
plt.show()
plt.boxplot(data);
```

### 4. Aggregate Functions:
#### Aggregate functions, often referred to as reduction functions,
#### are operations that perform a computation on an array and return a single result. 
#### These functions aggregate or reduce the values in the array to a single value based on a specific operation.
#### Examples are;

#### np.sum :Computes the sum of array elements usually along a specified axis(rows,columns)


```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
result = np.sum(arr)
print(result)
```

#### np.mean:Calculates the mean or average of array elements 

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
result = np.mean(arr)
print(result) 
```

#### np.median: Calculates the median value of array elements along a specified axis.

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
result = np.median(arr)
print(result)
```

#### np.min and np.max: Find the minimum and maximum values in an array along a specified axis.

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
min_val = np.min(arr)
max_val = np.max(arr)
print(min_val,max_val)
```

#### np.prod: Computes the product of array elements along a specified axis.

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
result = np.prod(arr)
print(result)
```

### 5. Universal Functions: 
#### Universal functions perform element-wise operations on the elements of one or more arrays,
#### producing an array as output.
#### Universal functions performs the following operations;
#### (a) Element wise operations
#### (b)Broadcasting
#### (c)Wide range functions

```python
#Broadcasting:
import numpy as np
arr1 = np.array([1, 2, 3])
scalar = 2
result = np.multiply(arr1, scalar)
print(result)
```

### 6. Difference between aggregate and universal functions.
#### Aggregate Functions are also called reduction functions, are used to compute a summary statistic or an aggregated value from an array or a specific axis of an array.
#### Examples like np.sum, np.mean, np.median, np.min, np.max, and np.std are aggregate functions.
#### They provide a single value that summarizes the information in the array.
#### They are used to obtain a single result that characterizes the distribution or properties of the data.
#### Universal functions operate element-wise on arrays,
#### performing element-wise operations on the corresponding elements of one or more arrays.
#### Examples are np.add, np.subtract, np.multiply, np.sin, np.exp,
#### and many others are universal functions. They are applied to each element of the array independently.
